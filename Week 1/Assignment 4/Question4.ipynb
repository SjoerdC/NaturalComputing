```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def random_sequence(n): return [np.random.binomial(1,0.5) for i in range(n)]
```

```python
def count_ones(seq): return (0 if len(seq) == 0 else (int(seq[0])+count_ones(seq[1:])))
```

```python
def transformed_sequence(seq,p): return [abs(s-1) if np.random.binomial(1,p) else s for s in seq]
```

```python
def counting_ones(n,iterations,p):   
    max_scores=[]
    # create a random sequence
    seq = random_sequence(n)
    # count the number of ones in the sequence
    max_score=count_ones(seq)
    # iterate through the number of iterations
    for i in range(iterations):
        # invert a bit with probability p
        new_seq = transformed_sequence(seq,p)
        # count number of ones in the transformed sequence
        score = count_ones(new_seq)
        # compare old and new sequence
        if score > max_score:
            max_score=score
            seq = new_seq
        max_scores.append(max_score)
        # stop when count reaches n
        if score == n:
            break
    while len(max_scores)<iterations:
        max_scores.append(n)
    # plot the fitness over the number of iterations
    plt.plot(range(0,iterations),max_scores)
    plt.xlabel("Number of Iterations")
    plt.ylabel("Fitness")
    plt.title("Fitness of the Genetic Algorithm over several iterations")
    plt.show()
    return max_scores,seq
```

```python
iterations = 1500
scores, seq = counting_ones(100,iterations,1/100)
```

```python
iterations = 10
scores, seq = counting_ones(100,iterations,1/100)
```
